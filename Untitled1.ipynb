```python

# SMART EXPENSE TRACKER APPLICATION

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime
import random
import os

plt.style.use("ggplot")
```

```python

#  CREATE SAMPLE DATASET


import random
import pandas as pd
from datetime import datetime, timedelta

categories = [
    "Food",
    "Transport",
    "Shopping",
    "Entertainment",
    "Medical",
    "Education",
    "Travel",
    "Utilities",
    "Other"
]

descriptions = {
    "Food": ["Lunch", "Dinner", "Breakfast", "Snacks", "Restaurant"],
    "Transport": ["Bus Ticket", "Taxi", "Fuel", "Train", "Metro"],
    "Shopping": ["Clothes", "Shoes", "Accessories", "Groceries", "Electronics"],
    "Entertainment": ["Movie", "Games", "Netflix", "Concert", "Park"],
    "Medical": ["Medicine", "Doctor Visit", "Hospital", "Checkup"],
    "Education": ["Books", "Course", "Stationery", "Exam Fees"],
    "Travel": ["Hotel", "Flight", "Tour", "Vacation"],
    "Utilities": ["Electricity", "Internet", "Water Bill", "Gas Bill"],
    "Other": ["Gift", "Donation", "Miscellaneous"]
}

start_date = datetime(2026, 1, 1)

expense_list = []

for i in range(100):

    random_date = start_date + timedelta(days=random.randint(0, 180))

    category = random.choice(categories)

    amount = random.randint(100, 5000)

    description = random.choice(descriptions[category])

    expense_list.append([
        random_date.strftime("%Y-%m-%d"),
        amount,
        category,
        description
    ])

df = pd.DataFrame(
    expense_list,
    columns=["Date", "Amount", "Category", "Description"]
)

df.to_csv("expenses.csv", index=False)

print("expenses.csv created successfully!")
print()
print(df.head())
```

```python

#  LOAD, EXPLORE AND CLEAN DATA


# Load Dataset
df = pd.read_csv("expenses.csv")

print("SMART EXPENSE TRACKER DATASET")


# Display first five records
print("\nFirst 5 Records")
print(df.head())

# Display last five records
print("\nLast 5 Records")
print(df.tail())

# Dataset Information
print("\nDataset Information")
df.info()

# Dataset Shape
print("\nDataset Shape")
print("Rows :", df.shape[0])
print("Columns :", df.shape[1])

# Column Names
print("\nColumn Names")
print(df.columns.tolist())

# Check Missing Values
print("\nMissing Values")
print(df.isnull().sum())

# Count Duplicate Records
print("\nDuplicate Records :", df.duplicated().sum())

# Remove Duplicate Records
df.drop_duplicates(inplace=True)

# Convert Date Column to Datetime Format
df["Date"] = pd.to_datetime(df["Date"])

# Convert Amount Column to Numeric
df["Amount"] = pd.to_numeric(df["Amount"], errors="coerce")

# Remove Rows with Missing Values
df.dropna(inplace=True)

# Reset Index
df.reset_index(drop=True, inplace=True)

print("\nData Cleaned Successfully!")

# Statistical Summary
print("\nStatistical Summary")
print(df.describe(include="all"))

# Save Cleaned Dataset
df.to_csv("expenses.csv", index=False)

print("\nCleaned dataset saved successfully.")
```

```python
class ExpenseTracker:

    # Constructor
    def __init__(self, filename="expenses.csv"):
        self.filename = filename
        self.load_data()

    # Destructor
    def __del__(self):
        print("Expense Tracker Closed Successfully.")

    # Load Data
    def load_data(self):
        try:
            self.df = pd.read_csv(self.filename)

            self.df["Date"] = pd.to_datetime(
                self.df["Date"],
                format="mixed",
                errors="coerce"
            )

            self.df["Amount"] = pd.to_numeric(
                self.df["Amount"],
                errors="coerce"
            )

            print("Data Loaded Successfully.")

        except FileNotFoundError:
            print("CSV File Not Found!")

            self.df = pd.DataFrame(columns=[
                "Date",
                "Amount",
                "Category",
                "Description"
            ])

    # Save Data
    def save_data(self):
        self.df.to_csv(self.filename, index=False)
        print("Data Saved Successfully.")

    # Add Expense
    def add_expense(self, date, amount, category, description):

        if amount <= 0:
            print("Amount must be greater than zero.")
            return

        new_expense = pd.DataFrame({
            "Date":[pd.to_datetime(date)],
            "Amount":[amount],
            "Category":[category],
            "Description":[description]
        })

        self.df = pd.concat(
            [self.df, new_expense],
            ignore_index=True
        )

        self.save_data()

        print("Expense Added Successfully.")
            # View All Expenses
    def view_expenses(self):

        if self.df.empty:
            print("No Expenses Available.")
        else:
            print(self.df)

    # Delete Expense
    def delete_expense(self, index):

        if 0 <= index < len(self.df):

            self.df.drop(index, inplace=True)
            self.df.reset_index(drop=True, inplace=True)

            self.save_data()

            print("Expense Deleted Successfully.")

        else:
            print("Invalid Index.")

    # Update Expense
    def update_expense(self, index, date, amount, category, description):

     if 0 <= index < len(self.df):

        self.df.loc[index, "Date"] = pd.to_datetime(date)
        self.df.loc[index, "Amount"] = amount
        self.df.loc[index, "Category"] = category
        self.df.loc[index, "Description"] = description
        self.df.loc[index, "Month"] = pd.to_datetime(date).strftime("%B")

        self.save_data()
        print("Expense Updated Successfully.")

     else:
        print("Invalid Index.")



    # Search by Category
    def search_category(self, category):

        result = self.df[
            self.df["Category"].str.lower() == category.lower()
        ]

        if result.empty:
            print("No Records Found.")
        else:
            print(result)

    # Search by Date
    def search_date(self, date):

        date = pd.to_datetime(date)

        result = self.df[
            self.df["Date"] == date
        ]

        if result.empty:
            print("No Records Found.")
        else:
            print(result)

    # Total Expense
    def total_expense(self):

        total = self.df["Amount"].sum()

        print(f"Total Expense : ₹{total:.2f}")

        return total

    # Average Expense
    def average_expense(self):

        average = self.df["Amount"].mean()

        print(f"Average Expense : ₹{average:.2f}")

        return average

    # Maximum Expense
    def maximum_expense(self):

        maximum = self.df["Amount"].max()

        print(f"Maximum Expense : ₹{maximum:.2f}")

        return maximum

    # Minimum Expense
    def minimum_expense(self):

        minimum = self.df["Amount"].min()

        print(f"Minimum Expense : ₹{minimum:.2f}")

        return minimum
            # ------------------------------------------------------
    # Expense Summary
    # ------------------------------------------------------
    def get_summary(self):

        print("=" * 50)
        print("EXPENSE SUMMARY")
        print("=" * 50)

        print(f"Total Transactions : {len(self.df)}")
        print(f"Total Expense      : ₹{self.df['Amount'].sum():.2f}")
        print(f"Average Expense    : ₹{self.df['Amount'].mean():.2f}")
        print(f"Maximum Expense    : ₹{self.df['Amount'].max():.2f}")
        print(f"Minimum Expense    : ₹{self.df['Amount'].min():.2f}")

    # ------------------------------------------------------
    # NumPy Analysis
    # ------------------------------------------------------
    def numpy_analysis(self):

        amounts = self.df["Amount"].to_numpy()

        print("\nNumPy Analysis")
        print("-" * 40)

        print("Mean :", np.mean(amounts))
        print("Median :", np.median(amounts))
        print("Maximum :", np.max(amounts))
        print("Minimum :", np.min(amounts))
        print("Standard Deviation :", np.std(amounts))
        print("Variance :", np.var(amounts))

    # ------------------------------------------------------
    # Category Report
    # ------------------------------------------------------
    def category_report(self):

        print("\nCategory Wise Report")
        print("-" * 40)

        report = self.df.groupby("Category")["Amount"].sum()

        print(report)

        return report

    # ------------------------------------------------------
    # Monthly Report
    # ------------------------------------------------------
    def monthly_report(self):

        self.df["Date"] = pd.to_datetime(
            self.df["Date"],
            format="mixed",
            errors="coerce"
        )

        self.df["Month"] = self.df["Date"].dt.month_name()

        report = self.df.groupby("Month")["Amount"].sum()

        print("\nMonthly Report")
        print("-" * 40)

        print(report)

        return report

    # ------------------------------------------------------
    # Monthly Average
    # ------------------------------------------------------
    def monthly_average(self):

        self.df["Date"] = pd.to_datetime(
            self.df["Date"],
            format="mixed",
            errors="coerce"
        )

        self.df["Month"] = self.df["Date"].dt.month_name()

        average = self.df.groupby("Month")["Amount"].mean()

        print("\nMonthly Average")
        print("-" * 40)

        print(average)

        return average

    # ------------------------------------------------------
    # Category Percentage
    # ------------------------------------------------------
    def category_percentage(self):

        total = self.df["Amount"].sum()

        percentage = (
            self.df.groupby("Category")["Amount"].sum()
            / total
        ) * 100

        print("\nCategory Percentage")
        print("-" * 40)

        print(percentage)

        return percentage

    # ------------------------------------------------------
    # Pivot Table
    # ------------------------------------------------------
    def create_pivot_table(self):

        self.df["Date"] = pd.to_datetime(
            self.df["Date"],
            format="mixed",
            errors="coerce"
        )

        self.df["Month"] = self.df["Date"].dt.month_name()

        pivot = pd.pivot_table(
            self.df,
            values="Amount",
            index="Category",
            columns="Month",
            aggfunc=np.sum,
            fill_value=0
        )

        print("\nPivot Table")
        print("-" * 40)

        print(pivot)

        return pivot

    # ------------------------------------------------------
    # Sort Expenses
    # ------------------------------------------------------
    def sort_expenses(self):

        sorted_df = self.df.sort_values(
            by="Amount",
            ascending=False
        )

        print("\nSorted Expenses")
        print(sorted_df)

        return sorted_df

    # ------------------------------------------------------
    # Filter Category
    # ------------------------------------------------------
    def filter_category(self, category):

        filtered = self.df[
            self.df["Category"] == category
        ]

        print(filtered)

        return filtered

    # ------------------------------------------------------
    # Filter Amount
    # ------------------------------------------------------
    def filter_amount(self, amount):

        filtered = self.df[
            self.df["Amount"] >= amount
        ]

        print(filtered)

        return filtered

    # ------------------------------------------------------
    # Filter Date
    # ------------------------------------------------------
    def filter_date(self, month):

        filtered = self.df[
            self.df["Date"].dt.strftime("%Y-%m") == month
        ]

        print(filtered)

        return filtered
            # ------------------------------------------------------
    # Bar Chart
    # ------------------------------------------------------
    def bar_chart(self):

        category_total = self.df.groupby("Category")["Amount"].sum()

        plt.figure(figsize=(10,6))
        plt.bar(category_total.index, category_total.values)

        plt.title("Category Wise Expenses")
        plt.xlabel("Category")
        plt.ylabel("Amount (₹)")
        plt.xticks(rotation=45)

        plt.tight_layout()
        plt.show()

    # ------------------------------------------------------
    # Pie Chart
    # ------------------------------------------------------
    def pie_chart(self):

        category_total = self.df.groupby("Category")["Amount"].sum()

        plt.figure(figsize=(8,8))

        plt.pie(
            category_total,
            labels=category_total.index,
            autopct="%1.1f%%",
            startangle=90
        )

        plt.title("Expense Distribution")

        plt.show()

    # ------------------------------------------------------
    # Line Chart
    # ------------------------------------------------------
    def line_chart(self):

        self.df["Date"] = pd.to_datetime(
            self.df["Date"],
            format="mixed",
            errors="coerce"
        )

        monthly = self.df.groupby(
            self.df["Date"].dt.strftime("%b")
        )["Amount"].sum()

        plt.figure(figsize=(10,6))

        plt.plot(
            monthly.index,
            monthly.values,
            marker="o"
        )

        plt.title("Monthly Expenses")
        plt.xlabel("Month")
        plt.ylabel("Amount (₹)")
        plt.grid(True)

        plt.show()

    # ------------------------------------------------------
    # Histogram
    # ------------------------------------------------------
    def histogram(self):

        plt.figure(figsize=(10,6))

        plt.hist(
            self.df["Amount"],
            bins=10,
            edgecolor="black"
        )

        plt.title("Expense Amount Distribution")
        plt.xlabel("Amount")
        plt.ylabel("Frequency")

        plt.show()

    # ------------------------------------------------------
    # Box Plot
    # ------------------------------------------------------
    def box_plot(self):

        plt.figure(figsize=(10,6))

        sns.boxplot(
            x="Category",
            y="Amount",
            data=self.df
        )

        plt.xticks(rotation=45)

        plt.title("Box Plot")

        plt.show()

    # ------------------------------------------------------
    # Count Plot
    # ------------------------------------------------------
    def count_plot(self):

        plt.figure(figsize=(10,6))

        sns.countplot(
            x="Category",
            data=self.df
        )

        plt.xticks(rotation=45)

        plt.title("Count Plot")

        plt.show()
```
